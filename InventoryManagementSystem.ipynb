```python
import json
import time

print("\n---------------------------------------------\n\t\tInventory Management System\n---------------------------------------------\n")
print("Chose the Function you want to perform:\n1.ADD ITEMS TO THE INVENTORY\n2.PURCHASE ITEMS FROM THE INVENTORY")
inpt = int(input("Enter 1 or 2 to proceed: "))

if(inpt == 1):

    itemNumber = int(input("Enter the number of Items that you want to add: "))
    fd = open("record.json", 'r')
    r = fd.read()
    fd.close()
    records = json.loads(r)

    for i in range(itemNumber):
        prod_id = str(input("Enter Product ID: "))
        if (prod_id in records.keys()):
            print("Product Already Exists!")
            qn = int(input("Enter quantity to be added: "))
            if(qn>=0):
                records[prod_id]['qn'] = records[prod_id]['qn'] + qn
            else:
                print("Invalid Quantity.")
        else:
            name = str(input("Enter name: "))
            pr = int(input("Enter price: "))
            qn = int(input("Enter quantity: "))
            records[prod_id] = {'name': name, 'pr': pr, 'qn': qn}
            print("New product added to inventory")
        js = json.dumps(records)
        fd = open("record.json", 'w')
        fd.write(js)
        fd.close()


elif(inpt == 2):
    print("\nINVENTORY:")
    fd = open("record.json", 'r')
    r = fd.read()
    fd.close()
    records = json.loads(r)
    price = 0
    keyValue = 0
    for key, value in records.items():
        print(key,":", value['name'], "- RS:", value['pr'])
    itemNumber = int(input("Enter the number of Items that you want to Purchase: "))
    sales = {}
    for i in range(itemNumber):
        iprice = 0
        prod_id = str(input("Enter Product ID: "))
        qn = int(input("Enter quantity: "))
        price = price + (records[prod_id]['pr'] *qn)
        iprice = (records[prod_id]['pr'] * qn)

        if (records[prod_id]['qn'] >= qn):
            records[prod_id]['qn'] = records[prod_id]['qn'] - qn
        else:
            print("Quantity not available")
            exit()

        keyValue = keyValue + 1
        itime = time.ctime()
        sales[keyValue] = {'prod' : prod_id, 'qn' : qn, 'amount': iprice, 'time of purchase' : itime}

        sale = json.dumps(sales)
        fdsale = open("sales.json", 'w')
        fdsale.write(sale)
        fdsale.close()

        js = json.dumps(records)
        fd = open("record.json", 'w')
        fd.write(js)
        fd.close()


    print("Billing Amount: ", price)
```
